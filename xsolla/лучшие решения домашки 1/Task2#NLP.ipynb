```python
import numpy as np
import pandas as pd 
```

```python
client = pd.read_csv('./data/clientSMS_v2.csv', sep='\t', names = ['text', 'sentiment'])
negative = pd.read_csv('./data/negative.csv', sep=';', usecols=[3], names=['text'])
negative['sentiment'] = 2
positive = pd.read_csv('./data/positive.csv', sep=';', usecols=[3], names=['text'])
positive['sentiment'] = 1
```

```python
kaggle_train = pd.read_json('./data/kaggle/train.json', encoding='UTF-8')
kaggle_train = kaggle_train[['text','sentiment']]
```

```python
print(positive.shape)
print(negative.shape)
print(client.shape)
print(kaggle_train.shape)
```

```python
df = pd.concat([kaggle_train, client, negative, positive], axis=0, ignore_index=True)
df['text'] = df['text'].apply(lambda x: x.strip())
```

```python
df['exclamations'] = df['text'].apply(lambda x: x.count('!'))
df['questions'] = df['text'].apply(lambda x: x.count('?'))
df['dots'] = df['text'].apply(lambda x: x.count('.'))
```

```python
df1 = df.sort_values(by=['text'], ascending=True)
for i in range(3):
    print(df1.iloc[i, 0])
    print()
    
print(df['exclamations'].sum(), df['questions'].sum(), df['dots'].sum())
```
